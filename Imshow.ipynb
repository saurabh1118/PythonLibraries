```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Reading the Image using matplotlib


```python
img = mpimg.imread('/content/ganesh.jpg')
```

```python
plt.imshow(img)
plt.show()
```

Reading image using opencv


```python
import cv2
```

```python
imgcv2 = cv2.imread('/content/ganesh.jpg')
```

```python
plt.imshow(imgcv2)
plt.show()
```

```python
imgcv2 = cv2.cvtColor(imgcv2 , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgcv2)
plt.show()
```

Changing Aspect = radio of image

```python
plt.imshow(img , aspect = 1)
plt.show()
```

```python
plt.imshow(img , aspect = 0.5)
plt.show()
```
